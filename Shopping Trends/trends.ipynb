```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("shopping_trends_updated.csv")
df.head()
```

Distribution of Customer Ages

```python
fig, ax = plt.subplots()
ax.hist(df['Age'], edgecolor='black')  # Your code here

ax.set_xlabel('Age')
ax.set_ylabel('Number of Customers')
ax.set_title('Distribution of Customer Ages')
```

Distribution of Customer Gender

```python
# Counting the occurrences of each gender
gender_counts = df['Gender'].value_counts()

# Creating the bar chart using Axes
fig, ax = plt.subplots()
ax.bar(gender_counts.index, gender_counts)

# Setting the labels and title
ax.set_xlabel('Gender')
ax.set_ylabel('Number of Customers')
ax.set_title('Gender Distribution of Customers')
```

Distribution of Items Purchased

```python
top_items = df['Item Purchased'].value_counts().head(10)
fig, ax = plt.subplots()
ax.bar(top_items.index, top_items)
ax.set_xlabel('Item')
ax.set_ylabel('Number of Purchases')
ax.set_title('Top Purchased Items')
ax.set_xticklabels(top_items.index, rotation=45)
plt.tight_layout()
```

Purchase Distribution by Item Category

```python
category_counts = df['Category'].value_counts() # Your code here

fig, ax = plt.subplots()
ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
ax.set_title('Purchases Distribution by Category')
```

Average Purchase Amount by Item Category

```python
average_purchase = df.groupby('Category')['Purchase Amount (USD)'].mean() # Your code here

fig, ax = plt.subplots()
ax.bar(average_purchase.index, average_purchase) # Your code here

ax.set_xlabel('Category')
ax.set_ylabel('Average Purchase Amount (USD)')
ax.set_title('Average Purchase Amount by Category')
ax.set_xticklabels(average_purchase.index,rotation=45)
```

Purchase Distribution by Season

```python
fig, ax = plt.subplots()
SEASON = df["Season"].value_counts()

ax.bar(SEASON.index,SEASON.values)

ax.set_xticklabels(('Spring', 'Fall', 'Winter', 'Summer'))

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
    ax.tick_params(axis = 'both', labelsize = 15)
plt.xlabel('Season', weight = "bold", color = "#D71313", fontsize = 14, labelpad = 20)
plt.ylabel('Number of Occurrences', weight = "bold", color = "#D71313", fontsize = 14, labelpad = 20)
```
